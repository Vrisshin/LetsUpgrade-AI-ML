## Statistical tests (Mann-Whitney)

```python
from scipy.stats import mannwhitneyu, ttest_ind, pearsonr
df = data[['Attrition','Age','DistanceFromHome','Education','MonthlyIncome',
'NumCompaniesWorked', 'PercentSalaryHike','TotalWorkingYears', 'TrainingTimesLastYear',
'YearsAtCompany','YearsSinceLastPromotion', 'YearsWithCurrManager']]
df.head()
```

```python
df_attrisyes = df[df.Attrition == 'Yes']
df_attrisno = df[df.Attrition == 'No']

display(df_attrisyes.head())
display(df_attrisno.head())
```

```python
def mannWhitneyu(df_temp, df_temp1):
    for col in df_temp.columns[1:]:
        print(f'Attrition & {col}')
        
        stat, p = mannwhitneyu(df_temp[col], df_temp1[col])
        print(stat, p)
        check(p)

def check(p):
        if p < 0.05:
            print(f"As the p value ", p ," < than 0.05, the H0 is rejected and H1 is accepted\n")
            print("There is significant differences in attrition (Y) and attrition (N)\n\n")
        else:
            print(f"As the p value", p , "> than 0.05, the H0 is accepted and H1 is rejected\n")
            print("There is no significant differences in  attrition (Y) and attrition(N)\n\n ")
```

```python
print('mannwhitneyu:\n')
mannWhitneyu(df_attrisyes, df_attrisno)
```

## pearson test

```python
data.Attrition=data.Attrition.replace("No",0)
data.Attrition=data.Attrition.replace("Yes",1)
data.head()
```

```python
def At_corr(col):
    print(f"H0: There is no correlation between {col} and Attrition")
    print(f"H1: There is correlation between {col} and Attrition")
    
    stats, p = pearsonr(data.Attrition, data[col])
    
    print(f"\nP-value is {p} and R is {stats}")
    
    if p > 0.05:
        print("H0 is accepted as pvalue is greater than 0.05\n")
        print(f"There is no correlation between {col} and Attrition")
    else:
        print("H1 is accepted as pvalue is less than 0.05\n")
        if stats > 0:
            print(f"There is positive correlation between {col} and Attrition\n\n")
        else:
            print(f"There is negative correlation between {col} and Attrition\n\n")
        

cols = ['Age', 'DistanceFromHome', 'MonthlyIncome', 'NumCompaniesWorked',
        'PercentSalaryHike','TotalWorkingYears','TrainingTimesLastYear', 
        'YearsAtCompany', 'YearsSinceLastPromotion','YearsWithCurrManager']
for cl in cols:
    At_corr(cl)
```
